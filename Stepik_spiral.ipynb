```
## Variant 1

n = 6
mtx = [[0 for i in range(n)] for j in range(n)]
mtx[0] = list(range(1, n+1))

# n=8   end = 5
for i in range(1, int(n/2)+1):
    
    # коли i стартує з 1
    for j in range(i, n-i+1): # правий стовбець завжди йде донизу
        mtx[j][n - i]= 1 + mtx[j-1][n-i]

    for j in reversed(range(i-1, n-i)): # нижній рядок завжди йде ліворуч
        mtx[n-i][j] = 1 + mtx[n-i][j+1]
        
    for j in reversed(range(i, n-i)): # лівий стовпець завжди йде знизу догори
        mtx[j][i-1]= 1 + mtx[j+1][i-1]
    
    for j in range(i, n-i): # верхній рядок завжди йде праворуч
        mtx[i][j] = 1 + mtx[i][j-1]
    
           
for line in mtx:
    print(*line)
```

```
## Variant 2
n = 8
mtx = [[0 for i in range(n)] for j in range(n)]

for i in range(0, int(n/2)+1):
    ## Для заповнення рядків змінюємо другий індекс при сталому першому
    ## Для заповнення стовпців змінюємо перший індекс при сталому другому

    ## Кожен наступний заповнений рядок буде віджирати індекси
    ## Для стовпців один зверху і один знизу
    ## Для рядків ліворуч і праворуч по одному
    
    for j in range(i, n-i): # верхній рядок завжди йде праворуч
        mtx[i][j] = 1 + mtx[i][j-1]

    for j in range(i+1, n-i): # правий стовбець завжди йде донизу
        # print(i)
        mtx[j][n-1-i]= 1 + mtx[j-1][n-1-i]

    for j in reversed(range(i, n-1-i)): # нижній рядок завжди йде ліворуч
        mtx[n-1-i][j] = 1 + mtx[n-1-i][j+1]
        
    for j in reversed(range(i+1, n-1-i)): # лівий стовпець завжди йде знизу догори
        mtx[j][i]= 1 + mtx[j+1][i]
    
for line in mtx:
    print(*line)
```
